```python
import pandas as pd 
```

```python
data = pd.read_csv('homeprices.csv')
data 
```

```python
 data.replace(to_replace=['monroe township','west windsor','robinsville'],value=["Faysal","EL HeLmeya","Giza"],inplace=True)
```

```python
data 
```

```python
data.groupby('town').mean()[['price']]
```

```python
data.groupby('town').median()
```

```python
dummies = pd.get_dummies(data.town)
dummies 
```

```python
Merged = pd.concat([data,dummies],axis='columns')
Merged
```

```python
F=Merged.drop(['town'],axis='columns')
F
```

```python
F=F.drop(['Giza'],axis='columns')
F
```

```python
X=F.drop(['price'],axis='columns')
X
```

```python
Y=F.price
Y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Model = LinearRegression()
Model
```

```python
Model.fit(X,Y)
```

```python
Model.predict([[2000,0,0]])
```

```python
Model.score(X,Y)
```
